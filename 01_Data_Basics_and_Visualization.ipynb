```python
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt

(X_train, y_train), (X_test,y_test) = mnist.load_data()

print("Training images:", X_train.shape)
print("Training labels:", y_train.shape)
print("Test images:", X_test.shape)
print("Test labels:", y_test.shape)

print("Unique classes:", set(y_train))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,8))
for i in range(25):
  plt.subplot(5, 5, i + 1)
  plt.imshow(X_train[i], cmap='gray')
  plt.title(str(y_train[i]))
  plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
from tensorflow.keras.datasets import cifar10

(Xc_train, yc_train), (Xc_test, yc_test) = cifar10.load_data()

print("CIFAR-10 train shape:", Xc_train.shape)
print("CIFAR-10 label shape:", yc_train.shape)

for i in range(25):
  plt.imshow(Xc_train[i])
  plt.title("Class: " + str(yc_train[i][0]))
  plt.axis('off')
  plt.show()
```
