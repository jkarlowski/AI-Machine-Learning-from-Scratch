```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor
X=np.linspace(0,2*np.pi,200).reshape(-1,1)
hidden_sizes=[5,10,20,50]
functions={
    "sin(x)":np.sin(X).ravel(),
    "cos(x)":np.cos(X).ravel(),
    "x^2":(X**2).ravel()
}
for func_name,y in functions.items():
  plt.figure(figsize=(10,8))
  for i,hls in enumerate(hidden_sizes):
    model=MLPRegressor(hidden_layer_sizes=(hls,),max_iter=2000,random_state=35)
    model.fit(X,y)
    y_pred=model.predict(X)
    plt.subplot(2,2,i+1)
    plt.title(f"Function: {func_name} with {hls} hidden layers")
    plt.plot(X,y,label="Original curve")
    plt.plot(X,y_pred,label="Predicted curve",linestyle="--")
    plt.legend()
  plt.show()
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from tensorflow.keras.datasets import fashion_mnist

(X_f, y_f),_=fashion_mnist.load_data()
X_f=X_f.reshape(X_f.shape[0],-1)/255.0
X_train,X_test,y_train,y_test=train_test_split(X_f,y_f,random_state=35)
sizes=[32,64,128]
iterations=[5,10,20,50]
for i in iterations:
  for hls in sizes:
    model=MLPClassifier(hidden_layer_sizes=(hls,),max_iter=i,warm_start=True,random_state=35,tol=1e-3)
    model.fit(X_train, y_train)
    train_acc=accuracy_score(y_train,model.predict(X_train))
    test_acc=accuracy_score(y_test,model.predict(X_test))
    print(f"Train accuracy for {hls} size and max_iter={i}: {train_acc:.4f}")
    print(f"Test accuracy for {hls} size and max_iter={i}: {test_acc:.4f}")
```
