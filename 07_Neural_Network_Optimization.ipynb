```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_wine
import time
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

data=load_wine()
X_train,X_test,y_train,y_test=train_test_split(
    data.data,data.target,test_size=0.2,random_state=35)
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
sizes=[16,32,64,128]
iterations=[20,50,100]
for i in iterations:
  train_acct=[]
  test_acct=[]
  for hls in sizes:
    model=MLPClassifier(hidden_layer_sizes=(hls,),max_iter=i,random_state=35)
    start_time=time.time()
    model.fit(X_train_scaled, y_train)
    end_time=time.time()
    train_time=end_time-start_time
    train_acc=model.score(X_train_scaled,y_train)
    test_acc=model.score(X_test_scaled,y_test)
    train_acct.append(train_acc)
    test_acct.append(test_acc)
    print(f"Train accuracy for {hls}: {train_acc:.4f}")
    print(f"Test accuracy for {hls}: {test_acc:.4f}")
    print(f"Under {train_time} training time")
  plt.figure(figsize=(6,4))
  plt.title(f"Accuracy vs Training Time with {i} iterations")
  plt.xlabel("Hidden Layer Size")
  plt.ylabel("Accuracy")
  plt.plot(sizes,train_acct,label="Train Accuracy")
  plt.plot(sizes,test_acct,label="Test Accuracy",linestyle="--")
  plt.legend()
  plt.grid(True)
  plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_moons

X_moons,y_moons=make_moons(n_samples=500, noise=0.2, random_state=35)
X_train,X_test,y_train,y_test=train_test_split(X_moons,y_moons,test_size=0.2,random_state=35)
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
log_reg=LogisticRegression(max_iter=1000)
start_timereg=time.time()
log_reg.fit(X_train_scaled, y_train)
end_timereg=time.time()
train_timereg=end_timereg-start_timereg
log_reg_acc=log_reg.score(X_test_scaled, y_test)
print(f"Test Accuracy (Logistic Regression): {log_reg_acc}")
print(f"Under {train_timereg} training time")
test_accmlp=[]
for hls in sizes:
  model=MLPClassifier(hidden_layer_sizes=(hls,),max_iter=1000,random_state=35)
  start_timemlp=time.time()
  model.fit(X_train_scaled, y_train)
  end_timemlp=time.time()
  train_timemlp=end_timemlp-start_timemlp
  test_acc=model.score(X_test_scaled,y_test)
  test_accmlp.append(test_acc)
  print(f"Train accuracy for {hls}: {train_acc}")
  print(f"Test accuracy for {hls}: {test_acc}")
  print(f"Under {train_timemlp} training time")
plt.figure(figsize=(8, 5))
plt.plot(sizes, test_accmlp, marker='o', label="MLP Accuracy")
plt.axhline(y=log_reg_acc, color='r', linestyle='--', label="Logistic Regression Accuracy")
plt.title("Porównanie MLP vs Model Liniowy")
plt.xlabel("Hidden Layer Size")
plt.ylabel("Test Accuracy")
plt.legend()
plt.grid(True)
plt.show()
```
